## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import datetime
import pandas as pd
test_date=datetime.datetime.strptime("01-01-23","%d-%m-%y")
k=41
date_generated=pd.date_range(test_date,periods=k)
print(date_generated.strftime("%d-%m-%y"))
```
